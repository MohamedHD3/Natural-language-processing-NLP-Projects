```python
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Generate a synthetic dataset with outliers
np.random.seed(42)
normal_data = np.random.normal(loc=0, scale=1, size=(800, 2))
outliers = np.random.uniform(low=-10, high=10, size=(20, 2))
data = np.vstack([normal_data, outliers])

# Create a DataFrame for the dataset
df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])

# Visualize the dataset
plt.scatter(df['Feature1'], df['Feature2'], s=5)
plt.title('Synthetic Dataset with Outliers')
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.show()

# from sklearn.model_selection import GridSearchCV

# # Example of grid search
# param_grid = {'contamination': [0.01, 0.02, 0.03, 0.04]}
# model = IsolationForest(random_state=42)
# grid_search = GridSearchCV(model, param_grid, cv=5, scoring='f1')
# grid_search.fit(df)
# print('Best Contamination:', grid_search.best_params_['contamination'])

from sklearn.model_selection import cross_val_score

for contamination_value in [0.01, 0.02, 0.03, 0.04]:
    model = IsolationForest(contamination=contamination_value, random_state=42)
    model.fit(df)
    df['is_outlier'] = model.predict(df)
    # Visualize the results
    plt.scatter(df['Feature1'][df['is_outlier'] == -1], df['Feature2'][df['is_outlier'] == -1], s=50, color='red', label='Outlier')
    plt.scatter(df['Feature1'][df['is_outlier'] == 1], df['Feature2'][df['is_outlier'] == 1], s=5, color='blue', label='Normal')
    plt.title(f'Isolation Forest Anomaly Detection : contamination_value = {contamination_value}' )
    plt.xlabel('Feature1')
    plt.ylabel('Feature2')
    plt.legend()
    plt.show()
```
